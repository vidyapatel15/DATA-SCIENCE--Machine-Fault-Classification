```python
# perfrom some processing to store the data
training_data=[]
def create_training_data():
    for category in CATEGORIES:
        path=os.path.join(DATADIR, category)
        class_num=CATEGORIES.index(category)
        for audio in os.listdir(path):
            audio_array, sr_array=librosa.load(os.path.join(path,audio))
            audio_array = audio_array[0:1024]
            training_data.append([audio_array,class_num])
create_training_data()
```

```python
# check the length of training data and store it for further usage
print(len(training_data))
lenofaudio = len(training_data)
```

```python
# split x and y
X=[]
y=[]

for categories, label in training_data:
    X.append(categories)
    y.append(label)
```

```python
y = np.array(y)
```

```python
X = np.vstack(X)
X = np.array(X).reshape(lenofaudio,-1)
X.shape
```

```python
X = X.reshape(1000,32,32)
```

**Generate image**

```python
from PIL import Image
```

```python
# Process and save images
def save_images_from_audio(X):
    # create a writable folder with full path
    import os
    img_folder = '/content/drive/MyDrive/audio_images'  # or another folder where you have write access
    os.makedirs(img_folder, exist_ok=True) # If the folder doesnt exist, this will create it for you
    for i in range(len(X)):
        # Reshape and normalize the audio data (assume audio data is 1D)
        image_array = (X[i] * 255).astype(np.uint8)
        img = Image.fromarray(image_array)
        img = img.convert("L")  # Convert to grayscale

        # Save the image with a unique name based on the index
        img.save(os.path.join(img_folder, f"audio_segment_{i+1}.png"))
        print(f"Saved image: audio_segment_{i+1}.png")

# Save all images in the 'audio_cats' folder
save_images_from_audio(X)

# Print completion message
print(f"All images have been saved in: {img_folder}")
```
